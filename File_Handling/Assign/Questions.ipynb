```python
import csv

with open('amazon_jobs_dataset.csv') as file_obj :
    file_data = csv.DictReader(file_obj, skipinitialspace = True)
    file_list = list(file_data) 

month_killed = {}
for row in file_list :
    key = row['location']
    value = row['Title']
    
    if value != '' :
        value = int(float(value))
    else :
        value = 0
        
    if key in month_killed :
        month_killed[key] += value
    else :
        month_killed[key] = value
        
for key, value in month_killed.items(): 
    print (key, value) 
    
```

```python
import csv
file = 'amazon_jobs_dataset.csv'
Country_basic_qualification=[]
with open(file,'r',encoding='utf-8') as csv_obj:
    file_data=csv.DictReader(csv_obj)
    for row in file_data:
        Country_basic_qualification.append([row['location'],row['BASIC QUALIFICATIONS']])
  ##get java developer from diffrent country      
def getJavaDeveloper(arr):
    arrlist=[]
    for i in arr:
        if 'Java' in i[1]:
            arrlist.append(i[0].split(',')[0])
    return arrlist

def getC(arr):
    arrlist=[]
    for i in arr:
        if 'C++' in i[1]:
            arrlist.append(i[0].split(',')[0])
    return arrlist

def get_Python(arr):
    arrlist=[]
    for i in arr:
        if 'Python' in i[1]:
            arrlist.append(i[0].split(',')[0])
    return arrlist


java_dev = getJavaDeveloper(Country_basic_qualification)
c_dev = getC(Country_basic_qualification)
python_dev = get_Python(Country_basic_qualification)

def createDictionary(arr):
    dictionary={i:0 for i in set(arr)}
    for i in arr:
        dictionary[i]=dictionary.get(i)+1
    return dictionary

def convertDictToList(dictionary):
    arrlist=[]
    for i in dictionary:
        arrlist.append([dictionary.get(i),i])
    arrlist.sort(reverse=True)
    return arrlist

Java_dict=createDictionary(java_dev)
Java_list=convertDictToList(Java_dict)
c_dict=createDictionary(c_dev)
c_list=convertDictToList(c_dict)
p_dict=createDictionary(python_dev)
p_list=convertDictToList(p_dict)
print(Java_list,c_list,p_list)
```
