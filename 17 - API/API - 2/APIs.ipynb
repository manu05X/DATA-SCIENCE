```python
#Print all sub-breeds of hound by making a get request to given URL.
#URL : https://dog.ceo/api/breeds/list/all
#Print name of sub-breeds in different lines.

import requests
import json

respons = requests.get(" https://dog.ceo/api/breeds/list/all")
data = json.loads(respons.text)
all_breed = data['message']
hound = all_breed['hound']

for dog in hound:
    print(dog)
```

```python
#Print name of all the planets from this URL : https://swapi.co/api/planets/.
#You can check the documentation here : https://swapi.co/documentation.


import requests
import json

respons = requests.get("https://swapi.co/api/planets")
data = json.loads(respons.text)
planet = data['results']

for i in planet:
    print(i['name'])
```

```python
#Print name of all online courses offered by Coding Ninjas. You can fetch the data using their open API :
#https://codingninjas.in/api/v3/courses.

import requests
import json

respons = requests.get("https://codingninjas.in/api/v3/courses")
datas = json.loads(respons.text)

all_data = datas['data']
all_course = all_data['courses']

for i in all_course:
    if i['available_online']==True:
        print(i['name'])
```

```python
#Print name of all the Coding Ninjas events (given in past registered events) which have word “Workshop” in their name.
#Data can be fetched from this URL : https://codingninjas.in/api/v3/events

import requests
import json

respons = requests.get("https://codingninjas.in/api/v3/events")
datas = json.loads(respons.text)
all_data = datas['data']['past_registered_events']
for i in all_data:
    if 'Workshop' in i['name']:
                         print(i['name'])
```
