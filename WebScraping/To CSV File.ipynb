```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://books.toscrape.com/')
response
```

```python
data = BeautifulSoup(response.text,'html.parser')
print(data.prettify())
```

* 1> Extract the URL of 1st Book
* 2> Load the html of 1st book page

```python
book_1 = data.find(class_ = 'product_pod')
book_1
```

```python
print(book_1.h3.a)
print(book_1.h3.a['href'])
```

```python
base_url = 'http://books.toscrape.com/'
b1_url = base_url + book_1.h3.a['href']
print(b1_url)
```

```python
response = requests.get(b1_url)
data = BeautifulSoup(response.text,'html.parser')
data
```

```python
title = data.h1.string
title
```

```python
price = data.find(class_ = 'price_color').string
price
```

```python
qty = data.find(class_ = 'instock availability')
qty
```

```python
qty = data.find(class_ = 'instock availability')
quantity = qty.contents[-1].strip()
quantity
```

```python
print(b1_url)
print(title)
print(price)
print(quantity)
```

```python
import re
```

```python
#re.serch take 2 argument as input e.g = re.search(pattern,string)

re.search('\d',quantity)
```

```python
quantity1 = int(re.search('\d+',quantity).group())
quantity1
```

```python
price1 = float(re.search('[\d.]+',price).group())
print(price1)
print(quantity1)
```

```python
book_details = []
book_details.append([title, b1_url, price1, quantity1])
book_details.append([title, b1_url, price1, quantity1])
book_details
```
