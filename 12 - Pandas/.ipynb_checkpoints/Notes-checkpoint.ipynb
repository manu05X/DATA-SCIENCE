##### Count of Flower
Q>

Find and print count of each kind of flower (separated by space)?

Note: Get the dataset from ''https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data''.

###### Feature Name are SepalLength,SepalWidth PetalLength, PetalWidth and Species

Print the count as Integer Value.

Output Format:

    count1 count2 count3 .....

```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd
col_name = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=col_name)
for i in iris['Species'].value_counts():
    print(i,end=" ")
```

##### Iris Virginica
Q>

Find the data of flower “Iris-virginica” type where petal-length > 1.5?

Note: Get the dataset from ''https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data''.

###### Feature Name are SepalLength,SepalWidth PetalLength, PetalWidth and Species

Print the count as Integer Value.

Output Format:

    feature1 feature2 feature3 feature4 feature5
    feature1 feature2 feature3 feature4 feature5
    feature1 feature2 feature3 feature4 feature5
    feature1 feature2 feature3 feature4 feature5
    . . .
    . . .
    . . .

```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd
col_name = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=col_name)
iris = iris[iris['Species'] == 'Iris-virginica'] 
iris = iris[iris['PetalLength'] >1.5]
res= iris.values

for i in range(res.shape[0]):
    for j in range(res.shape[1]):
        print(res[i][j], end = " ")
    print()
```

##### Iris Values
Q>

Find and print the minimum, maximum and average value of the feature for each kind of flower ?

Note: Get the dataset from ''https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data''.

###### Feature Name are SepalLength,SepalWidth PetalLength, PetalWidth and Species

Print the value with two decimal places.

Note: Order for flower is Iris-setosa, Iris-versicolor and Iris-virginica.

Output Format:

    minSL minSW minPL minPW Iris-setosa
    maxSL maxSW maxPL maxPW Iris-setosa
    avgSL avgSW avgPL avgPW Iris-setosa
    . . .
    . . .
    . . .

```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd
col_name = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=col_name)

sp1 = 'Iris-setosa'
sp2 = 'Iris-versicolor'
sp3 = 'Iris-virginica'

spdetail1 = df[df['Species']==sp1].describe()
spdetail2 = df[df['Species']==sp2].describe()
spdetail3 = df[df['Species']==sp3].describe()

#print the detail of 1st species
for i in spdetail1.iloc[[3]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp1)
for i in spdetail1.iloc[[7]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp1)
for i in spdetail1.iloc[[1]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp1)

#print the detail of 2nd species
for i in spdetail2.iloc[[3]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp2)
for i in spdetail2.iloc[[7]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp2)
for i in spdetail2.iloc[[1]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp2)

#print the detail of 3rd species
for i in spdetail3.iloc[[3]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp3)
for i in spdetail3.iloc[[7]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp3)
for i in spdetail3.iloc[[1]].values[0]:
    print(format(i, '.2f'), end=" ")
print(sp3)
```
